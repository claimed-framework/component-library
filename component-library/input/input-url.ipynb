```python
import logging
import os
import re
import requests
import sys
import wget
```

```python
# path and file name for output
output_data = os.environ.get('output_data', 'data.csv')

# url of souce
url = os.environ.get('url')

# temporal data storage for local execution
data_dir = os.environ.get('data_dir', '../../data/')
```

```python
parameters = list(
    map(lambda s: re.sub('$', '"', s),
        map(
            lambda s: s.replace('=', '="'),
            filter(
                lambda s: s.find('=') > -1 and bool(re.match(r'[A-Za-z0-9_]*=[.\/A-Za-z0-9]*', s)),
                sys.argv
            )
    )))

for parameter in parameters:
    logging.debug('Parameter: ' + parameter)
    exec(parameter)
```

```python
header={"user-agent":"Mozilla/5.0 (Windows NT 6.0) Gecko/20100101 Firefox/14.0.1"}
r = requests.get(url, headers=header)
print(output_data)
logging.debug('output_data: ' + output_data)
target_filename = url.split('/')[-1:][0]
destination = os.path.join(output_data) + '/' + target_filename
os.remove(destination) if os.path.exists(destination) else None
with open(destination, "wb") as f:
    f.write(r.content)
```
